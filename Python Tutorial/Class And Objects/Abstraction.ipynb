```python
from abc import ABC,abstractmethod

## Abstract base cclass

class Vehicle(ABC):
    def drive(self):
        print("The vehicle used for driving")

    @abstractmethod
    def start_engine(self):
        pass

class Car(Vehicle):
     def start_engine(self):
        print("Car Engine started")

def operate_vehicle(vehicle):
    vehicle.start_engine()
    vehicle.drive()
    
car=Car()
operate_vehicle(car)
```

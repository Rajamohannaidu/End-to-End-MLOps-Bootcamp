```python
##accessing Dictionary elements

student={"name":"Raja","age":42,"grade":'A'}
print(student)
```

```python
print(student['grade'])
print(student['age'])
```

```python
##accessing using get

print(student.get('grade'))

print(student.get('last_name'))

print(student.get('last_name',"Not Available"))
```

```python
##Modifying dictionary elements
##Dictonary are mutable

print(student)
```

```python
student["age"]=33  ##updated
print(student)
student["address"]="India" ##added
print(student)
```

```python
del student['grade'] ##delete
print(student)
```

```python
##Dictionary methods

keys=student.keys()  ##get all the keys
print(keys)
values=student.values() ##get all the values
print(values)

items=student.items()
print(items)
```

```python
## shallow copy
student_copy=student
print(student)
print(student_copy)
```

```python
student["name"]="Raja2"

print(student)
print(student_copy)
```

```python
student_copy1=student.copy() ##shallow copy

print(student_copy1)

print(student)
```

```python
student["name"]="Raja3"

print(student)
print(student_copy1)
```

```python
##Iterating over Dictionaries

## you can use loops to iterate over dictionaries,keys,values or items

for keys in student.keys():
    print(keys)
```

```python
for values in student.values():
    print(values)
```

```python
for key,value in student.items():
    print(f"{key}:{value}")
```

```python
## Nested dictionaries

students = {
    "student1":{"name":"krish","age":32},
    "student2":{"name":"Raja","age":42},
}

print(students)
```

```python
print(students["student2"]["name"])
print(students["student2"]["age"])
```

```python
students.items()
```

```python
for student_id,student_info in students.items():
    print(f"{student_id}:{student_info}")
    for key,value in student_info.items():
        print(f"{key}:{value}")
```

```python
## Dictionry Comprehension

squares={x:x**2 for x in range(5)}
print(squares)

```

```python
##Condition dictionary comprehension

evens = { x:x**2 for x in range(10) if x%2==0}
print(evens)
```

```python
##Practical Examples

##Use a dictionary to count the frequnecy of the elements in a list

numbers=[1,2,2,3,3,3,4,4,4,4]

frequency={}

for number in numbers:
    if number in frequency:
        frequency[number]+=1
    else:
        frequency[number]=1
print(frequency)
```

```python
##Merge 2 dictionaries

dict1={"a":1,"b":2}
dict2={"b":3,"c":4}

merged_dict={**dict1,**dict2}
print(merged_dict)
```
